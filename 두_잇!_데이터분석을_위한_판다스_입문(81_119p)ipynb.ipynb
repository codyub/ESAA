# 다변량 데이터로 다변량 그래프 그리기: 산점도 그래프
1.	산점도 그래프에 성별을 새 변수로 추가: 산점도 그래프에서 색상으로 구분
숫자에 성별 변수를 넣어야 하므로 문자열을 정수로 변경해야 함


```python
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 2
```

2.	데이터 프레임에 반환값을 추가

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

3.	인원수도 산점도 그래프에 추가: 점의 크기로 표현 (s = 점의 크기, c = 점의 색상, alpha = 투명도)

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter( x = tips['total_bill'], y = tips['tip'], s = tips['size'] * 10, c = tips['sex_color'], alpha = 0.5)
axes1.set_title('total bill vs tip colored by sex and sized by size')
axes1.set_xlabel('total bill')
axes1.set_ylabel('tip')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('total bill histogram with density plot')
```



3. displot 메서드의 밀집도 그래프를 제외하기 위한 옵션 : kde = False

# 다양한 종류의 이변량 그래프 그리기

1. seaborn 라이브러리로 산점도 그래프 그리기

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips) # 산점도 그래프와 회귀선을 함께 그려줌.
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False) # 회귀선 옵션 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

2. 산점도 그래프와 히스토그램을 한 번에 그리기: jointplot(x = '원하는 열', y = '원하는 열', data = 데이터프레임)

```python
joint = sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

3. 산점도 그래프가 점이 겹칠 때 구분하기 쉽게 함: 육각 그래프(hexbin,데이터 개수가 많으면 진한 색으로 표현됨)

```python
joint = sns.jointplot(x='total_bill',y='tip',data=tips, kind="hex")
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

4. 이차원 밀집도 그리기: kdeplot(data=원하는 열 값, data2=원하는 열값, shade=True(그래프에 음영효과 주기))

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

5. 바 그래프 그리기: 지정한 변수의 평균을 계산해 그림, barplot(x='열이름', y='열이름', data=데이터)

```python
ax = plt.subplots()
ax = sns.barplot(x='time',y='total_bill',data=tips)
ax.set_title('Bar plot of average total bill for tilme of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

6. 박스 그래프 그리기

```python
ax = plt.subplots()
ax = sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of average total bill for tilme of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

7. 바이올린 그래프: 박스그래프는 다양한 통계 수치 확인을 위해서는 좋지만, 데이터 분산이 모호하게 표현되므로 사용한다. 

```python
ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('Violin plot of average total bill for tilme of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

8. 관계 그래프 그리기 : 위 그래프들을 종합한 형태

```python
fig = sns.pairplot(tips)
```

9. 관계 그래프는 중복된 정보가 표현된다는 단점이 있다.(대각선을 기준으로 대칭) --> 중복된 그래프가 그려지는 위치를 직접 지정해 우너하는 그래프로 교체 가능



> map_upper : 대각선을 기준으로 위쪽에 그릴 그래프 지정

> map_lower : 대각선을 기준으로 아래에 그릴 그래프 지정

> map_diag : 대각선을 중심으로 그래프를 그림






```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.regplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```
